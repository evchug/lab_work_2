```python
import requests
from bs4 import BeautifulSoup
import re
from tqdm import tqdm
```

```python
urls = []
start_url = 'https://gazeta-vp.ru/list/page'
for x in range(1, 2516):
    url = start_url + '/' + str(x)
    urls.append(url)
#   url   
    
out = []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    for link in soup.find_all('a'):
        elem = link.get('href')
        if type(elem) == str:
            out.append(elem)
#        
            
newslinks = []
for link in out:
    if len(str(link))>60 and 'openvlz' not in link and '#respond' not in link:
        newslinks.append(link)
newslinks = list(dict.fromkeys(newslinks))
#      ,    

newslinks
```

```python
for link in newslinks:
    if '%' not in str(link) and 'tekush' not in str(link):
        rs = requests.get(link)
        root = BeautifulSoup(rs.content, 'html.parser')
        article = root.select_one('article')
        entry = '\n=====\n' + str(link) + '\n-----\n' + article.text
        with open('result.txt', 'a', encoding='utf-8') as file:
            file.write(entry)
    else:
        pass
```
